1. What is the difference between interpreted and compiled languages?

Interpreted languages execute code line by line (e.g., Python), while compiled languages are transformed into machine code before execution (e.g., C/C++), resulting in faster execution.



2. What is exception handling in Python?

Exception handling allows the program to deal with runtime errors using try, except, else, and finally blocks to prevent crashes and manage error recovery.



3. What is the purpose of the finally block in exception handling?

The finally block is used to execute code regardless of whether an exception occurred or not, typically for cleanup operations.



4. What is logging in Python?

Logging is the process of recording events or messages during program execution, often for debugging or monitoring purposes, using Pythonâ€™s logging module.



5. What is the significance of the _del_ method in Python?

The _del_ method is a destructor called when an object is about to be destroyed. It is used to define cleanup actions before the object is removed from memory.



6. What is the difference between import and from ... import in Python?

import module imports the whole module. from module import function imports specific attributes or functions directly.



7. How can you handle multiple exceptions in Python?

Use multiple except blocks or combine exceptions in a tuple:

try:
    # code
except (TypeError, ValueError) as e:
    # handle both



8. What is the purpose of the with statement when handling files in Python?

The with statement ensures that a file is properly closed after its suite finishes, even if an exception is raised.



9. What is the difference between multithreading and multiprocessing?

Multithreading runs threads in the same memory space; multiprocessing runs processes in separate memory spaces. Multiprocessing is better for CPU-bound tasks.



10. What are the advantages of using logging in a program?

Helps in debugging, monitoring application behavior, saving history of execution, and identifying issues after deployment.



11. What is memory management in Python?

It refers to how Python handles memory allocation and deallocation automatically using reference counting and garbage collection.



12. What are the basic steps involved in exception handling in Python?

Use try to enclose code, except to catch exceptions, optionally else to run if no exception, and finally for cleanup.



13. Why is memory management important in Python?

To optimize resource usage, improve performance, and prevent memory leaks or crashes due to exhausted memory.



14. What is the role of try and except in exception handling?

try defines a block where exceptions might occur. except catches and handles those exceptions gracefully.



15. How does Python's garbage collection system work?

Python uses reference counting and a cyclic garbage collector to detect and reclaim memory occupied by unused objects.



16. What is the purpose of the else block in exception handling?

The else block runs if no exceptions are raised in the try block, useful for code that should run only if the try was successful.



17. What are the common logging levels in Python?

DEBUG, INFO, WARNING, ERROR, and CRITICAL.



18. What is the difference between os.fork() and multiprocessing in Python?

os.fork() creates a child process in Unix systems by duplicating the parent, while multiprocessing is a cross-platform module to create processes.



19. What is the importance of closing a file in Python?

It releases system resources and ensures data is properly written to disk. Using with handles this automatically.



20. What is the difference between file.read() and file.readline() in Python?

file.read() reads the entire file or specified number of bytes. file.readline() reads a single line from the file.



21. What is the logging module in Python used for?

22. What is the os module in Python used for in file handling?

23. What are the challenges associated with memory management in Python?

```python
# 4. Write a pyhton script that reads from one file and writes its contant to another file.


# File Copy Script

# Specify the source and destination file paths
source_file = 'source.txt'
destination_file = 'destination.txt'

try:
    # Open the source file in read mode
    with open(source_file, 'r') as src:
        content = src.read()

    # Open the destination file in write mode and write the content
    with open(destination_file, 'w') as dest:
        dest.write(content)

    print("File content copied successfully.")

except FileNotFoundError:
    print(f"Error: The file '{source_file}' was not found.")

except Exception as e:
    print(f"An error occurred: {e}")
```

```python
# 18. How would you implement a basi loging setup that logs to a file with rotation after 1MB.

import logging
from logging.handlers import RotatingFileHandler

handler = RotatingFileHandler("rotated.log", maxBytes=1_000_000, backupCount=3)
logging.basicConfig(handlers=[handler], level=logging.INFO,
                    format="%(asctime)s - %(levelname)s - %(message)s")

logging.info("Starting the logging service.")
```
